```python
test = ['123 328  51 64 \n',
        ' 45 64  387 23 \n',
        '  6 98  215 314\n',
        '*   +   *   +  ']
```

```python
def parse_data(data, part=1):
    rows = []
    func = data[-1].strip().split()
    
    if part == 1:
        for line in data[:-1]:
            row = line.strip().split()
            for i, n in enumerate(row):
                row[i] = int(n)
            rows.append(row)
                
    elif part == 2:
        func = func[::-1]
        for i in range(0, len(data)-1):
            rows.append([])
            
        count = 0
        for i in range(len(data[0])-2, -1, -1):  
            num = ''
            for j in range(0, len(data)-1):
                num += data[j][i]
            num = num.strip().replace(' ', '0')
            
            if len(num) == 0:
                if count < len(rows):
                    if func[len(rows[0])-1] == '*':
                        extra = 1
                    elif func[len(rows[0])-1] == '+':
                        extra = 0
                        
                    while count < len(rows):
                        rows[count].append(extra)
                        count += 1
                count = 0
                
            else:
                rows[count].append(int(num))
                count += 1
            
    return rows, func

def do_homework(data, part=1):
    rows, func = parse_data(data, part)
    
    result = 0
        
    for i in range(0, len(rows[0])):
        if func[i] == '*':
            answer = rows[0][i]
            for row in rows[1:]:
                answer *= row[i]
                
        elif func[i] == '+':
            answer = rows[0][i]
            for row in rows[1:]:
                answer += row[i]

        result += answer
    return result
```

```python
print('Part 1 test:', do_homework(test), do_homework(test)==4277556)
```
