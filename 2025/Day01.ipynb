```python
test = ['L68',
        'L30',
        'R48',
        'L5',
        'R60',
        'L55',
        'L1',
        'L99',
        'R14',
        'L82']
```

```python
def turn_dial(sequence, dial=50):
    count_0 = 0
    point_0 = 0
    
    for s in sequence:
        lr = s[0]
        n = int(s[1:])

        if lr == 'L':
            dial -= n
            if dial < 0:
                point_0 += abs(dial)//100
                if dial + n != 0:
                    point_0 += 1
        else:
            dial += n
            if dial > 99:
                point_0 += dial//100

        if dial == 0:
            point_0 += 1

        dial %= 100

        if dial == 0:
            count_0 += 1

    return dial, count_0, point_0
```

```python
print(turn_dial(test))
```

```python
with open('input_day01.txt', 'r') as f:
    data = f.readlines()
    f.close()
print('Part 1 result:', turn_dial(data)[1])
```
