```python
import numpy as np
from PIL import Image
```

```python
def free_rolls(grid, prnt=False):

    moveable = 0
    for x in range(0, len(grid)):
        for y in range(0, len(grid[x])):
            if grid[x][y] == '.':
                continue
                
            neighbour_rolls = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    if dx == 0 and dy == 0:
                        continue
                    if x+dx < 0 or x+dx >= len(grid) or\
                        y+dy < 0 or y+dy >= len(grid[x]):
                        continue
                    if grid[x+dx][y+dy] != '.':
                        neighbour_rolls += 1

            if neighbour_rolls < 4:
                moveable += 1
                grid[x][y] = 'X'

    if prnt:
        for x in range(0, len(grid)):
            line = ''
            for y in range(0, len(grid[x])):
                line += grid[x][y]
            print(line)
        print()

    return moveable, grid

def make_frame(grid, frame_num):
    image = []
    for x in range(0, len(grid)):
        row = []
        for y in range(0, len(grid)):
            if grid[x][y] == '.':
                row.append([232,236,251])
            elif grid[x][y] == '@':
                row.append([25,101,176])
            elif grid[x][y] == 'X':
                row.append([220,5,12])
        image.append(row)

    image = np.array(image)
    frame = Image.fromarray(image.astype('uint8'), mode='RGB')
    frame = frame.resize((image.shape[0]*8,image.shape[1]*8), resample=Image.NEAREST)
    frame.save('./frames/Day04/day4_'+str(frame_num).zfill(5)+'.png')

    return frame_num+1


def remove_rolls(data, prnt=False):
    grid = []
    for line in data:
        grid.append(list(line.strip()))

    removed = 0
    frame_num = 0

    #Make first frame
    frame_num = make_frame(grid, frame_num)

    while True:
        movable, grid = free_rolls(grid, prnt)
        if movable == 0:
            break
        removed += movable

        #make frame
        frame_num = make_frame(grid, frame_num)

        #clean grid
        for x in range(0, len(grid)):
            for y in range(0, len(grid[x])):
                if grid[x][y] == 'X':
                    grid[x][y] = '.'

    #Make last frame
    frame_num = make_frame(grid, frame_num)

    return
```

```python
with open('input_day04.txt', 'r') as f:
    data = f.readlines()
    f.close()

remove_rolls(data)
```
