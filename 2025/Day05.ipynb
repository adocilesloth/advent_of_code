```python
def clean_ranges(data):
    remove_count = 1
    while remove_count > 0:
        remove_count = 0

        #Forward pass
        to_remove = []
        for i in range(0, len(data)-1):
            for j in range(i+1, len(data)):                
                if data[i][0] >= data[j][0] and data[i][0] <= data[j][1] and\
                   data[i][1] >= data[j][0] and data[i][1] <= data[j][1]:
                    if i not in to_remove:
                            to_remove.append(i)
                
                elif data[i][0] >= data[j][0] and data[i][0] <= data[j][1]:
                    data[j][1] = data[i][1]
                    if i not in to_remove:
                        to_remove.append(i)
    
                elif data[i][1] >= data[j][0] and data[i][1] <= data[j][1]:
                    data[j][0] = data[i][0]
                    if i not in to_remove:
                        to_remove.append(i)

        to_remove.sort()
        to_remove = to_remove[::-1]

        for i in to_remove:
            data.pop(i)
            remove_count += 1

        #Reverse pass
        #We will not remove the last range if it's inside an earlier range...
        #...so run it backwards
        to_remove = []
        for i in range(len(data)-1, 0, -1):
            for j in range(i-1, -1, -1):                
                if data[i][0] >= data[j][0] and data[i][0] <= data[j][1] and\
                   data[i][1] >= data[j][0] and data[i][1] <= data[j][1]:
                    if i not in to_remove:
                            to_remove.append(i)
                
                elif data[i][0] >= data[j][0] and data[i][0] <= data[j][1]:
                    data[j][1] = data[i][1]
                    if i not in to_remove:
                        to_remove.append(i)
    
                elif data[i][1] >= data[j][0] and data[i][1] <= data[j][1]:
                    data[j][0] = data[i][0]
                    if i not in to_remove:
                        to_remove.append(i)

        to_remove.sort()
        to_remove = to_remove[::-1]

        for i in to_remove:
            data.pop(i)
            remove_count += 1

    return data

def is_fresh(ranges, ingredient):
    for rng in ranges:
        if ingredient >= rng[0] and ingredient <= rng[1]:
            return True
    return False

def count_possible_fresh(ranges):
    possible = 0
    for rng in ranges:
        possible += (rng[1]-rng[0])+1
    return possible

def get_fresh(data):
    ranges = []
    ingredients = []

    is_range = True
    for line in data:       
        if '-' in line:
            rng = line.strip().split('-')
            try:
                rng[0] = int(rng[0])
            except:
                print(line)
                int(rng[0])
            rng[1] = int(rng[1])
            ranges.append(rng)
        elif is_range:
            is_range = False
        else:
            ingredients.append(int(line.strip()))

    ranges = clean_ranges(ranges)

    fresh_count = 0
    for ingredient in ingredients:
        fresh_count += is_fresh(ranges, ingredient)

    possible_count = count_possible_fresh(ranges)

    return fresh_count, possible_count
```
