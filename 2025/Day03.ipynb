```python
test = ['987654321111111',
        '811111111111119',
        '234234234234278',
        '818181911112111']
```

```python
def max_joltage(data):
    total = 0
    for line in data:
        bank = list(map(int, line))
        bat1 = max(bank[:-1])
        arg1 = bank.index(bat1)
        
        bank = bank[arg1+1:]
        bat2 = max(bank)

        joltage = int(str(bat1)+str(bat2))
            
        total += joltage
        #print(line, joltage)

    return total

print('Part 1 test:', max_joltage(test), max_joltage(test)==357)
```

```python
with open('input_day03.txt', 'r') as f:
    data = f.readlines()
    f.close()
data = [line.strip() for line in data]

print('Part 1 result:', max_joltage(data))
```

```python
def max_override_joltage(data):
    total = 0
    for line in data:
        bank = list(map(int, line))

        start = 0
        end = len(bank)-11
        joltage = ''
        
        while len(joltage) < 12:
            bat = max(bank[start:end])
            joltage += str(bat)
            start += bank[start:end].index(bat)+1
            end += 1
            
        joltage = int(joltage)

        total += joltage
        
    return total

print('Part 2 test:', max_override_joltage(test), max_override_joltage(test)==3121910778619)
```
