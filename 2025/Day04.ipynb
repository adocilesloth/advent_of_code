```python
test = ['..@@.@@@@.',
        '@@@.@.@.@@',
        '@@@@@.@.@@',
        '@.@@@@..@.',
        '@@.@@@@.@@',
        '.@@@@@@@.@',
        '.@.@.@.@@@',
        '@.@@@.@@@@',
        '.@@@@@@@@.',
        '@.@.@@@.@.']
```

```python
def free_rolls(grid, prnt=False):

    moveable = 0
    for x in range(0, len(grid)):
        for y in range(0, len(grid[x])):
            if grid[x][y] == '.':
                continue
                
            neighbour_rolls = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    if dx == 0 and dy == 0:
                        continue
                    if x+dx < 0 or x+dx >= len(grid) or\
                        y+dy < 0 or y+dy >= len(grid[x]):
                        continue
                    if grid[x+dx][y+dy] != '.':
                        neighbour_rolls += 1

            if neighbour_rolls < 4:
                moveable += 1
                grid[x][y] = 'X'

    if prnt:
        for x in range(0, len(grid)):
            line = ''
            for y in range(0, len(grid[x])):
                line += grid[x][y]
            print(line)
        print()

    return moveable, grid

def count_free_rolls(data, prnt=False):
    grid = []
    for line in data:
        grid.append(list(line.strip()))

    return free_rolls(grid, prnt)[0]

print('Part 1 test:', count_free_rolls(test, True), count_free_rolls(test)==13)
```

```python
with open('input_day04.txt', 'r') as f:
    data = f.readlines()
    f.close()

print('Part 1 result:', count_free_rolls(data))
```

```python
def remove_rolls(data, prnt=False):
    grid = []
    for line in data:
        grid.append(list(line.strip()))

    removed = 0

    while True:
        movable, grid = free_rolls(grid, prnt)
        if movable == 0:
            break
        removed += movable

        #clean grid
        for x in range(0, len(grid)):
            for y in range(0, len(grid[x])):
                if grid[x][y] == 'X':
                    grid[x][y] = '.'

    return removed

print('Part 2 test:', remove_rolls(test, True), remove_rolls(test)==43)
```

```python
print('Part 2 result:', remove_rolls(data))
```
