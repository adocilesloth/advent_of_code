```python
test = ['0:',
        '###',
        '##.',
        '##.',
        '',
        '1:',
        '###',
        '##.',
        '.##',
        '',
        '2:',
        '.##',
        '###',
        '##.',
        '',
        '3:',
        '##.',
        '###',
        '##.',
        '',
        '4:',
        '###',
        '#..',
        '###',
        '',
        '5:',
        '###',
        '.#.',
        '###',
        '',
        '4x4: 0 0 0 0 2 0',
        '12x5: 1 0 1 0 2 2',
        '12x5: 1 0 1 0 3 2']
```

```python
def parse_data(data):
    presents = {}
    regions = []
    
    for line in data:
        if len(line.strip()) == 0:
            continue
            
        elif len(line.strip()) == 2:
            idx = int(line[0])
            presents[idx] = []
            
        elif len(line.strip()) == 3:
            presents[idx].append([])
            for ch in line.strip():
                if ch == '.':
                    presents[idx][-1].append(0)
                elif ch == '#':
                    presents[idx][-1].append(1)
                    
        else:
            area, counts = line.strip().split(': ')
            
            area = area.split('x')
            for i, a in enumerate(area):
                area[i] = int(a)
            
            counts = counts.split(' ')
            for i, c in enumerate(counts):
                counts[i] = int(c)
                
            regions.append([area, counts])
            
    presents_area = {}
    for key in presents.keys():
        area = 0
        for i in range(0, len(presents[key])):
            for j in range(0, len(presents[key][i])):
                area += presents[key][i][j]
        presents_area[key] = area

    return presents, regions, presents_area

def get_regions_that_fit(presents, regions, presents_area):
    count = 0
    for region in regions:
        area = region[0][0] * region[0][1]
        num_area = (region[0][0]//3) * (region[0][1]//3)
        
        num_presents = 0
        present_area = 0
        
        for i, num in enumerate(region[1]):
            num_presents += num
            present_area += (num*presents_area[i])
            
        if num_presents < num_area:
            count += 1
            continue

        if area < present_area:
            continue
        count += 1
        
    return count

def run_parts(data, part=1):
    presents, regions, presents_area = parse_data(data)
    
    if part == 1:
        return get_regions_that_fit(presents, regions, presents_area)
    return
```

```python
print('Part 1 test:', run_parts(test), run_parts(test)==2)
```

```python
with open('input_day12.txt', 'r') as f:
    data = f.readlines()
    f.close()
    
print('Part 1 result:', run_parts(data))
```
