```python
from collections import deque
from copy import deepcopy
```

```python
test = ['aaa: you hhh',
        'you: bbb ccc',
        'bbb: ddd eee',
        'ccc: ddd eee fff',
        'ddd: ggg',
        'eee: out',
        'fff: out',
        'ggg: out',
        'hhh: ccc fff iii',
        'iii: out']

test2 = ['svr: aaa bbb',
         'aaa: fft',
         'fft: ccc',
         'bbb: tty',
         'tty: ccc',
         'ccc: ddd eee',
         'ddd: hub',
         'hub: fff',
         'eee: dac',
         'dac: fff',
         'fff: ggg hhh',
         'ggg: out',
         'hhh: out']
```

```python
def build_network(data):
    network = {}
    for line in data:
        inpu, outp = line.strip().split(': ')
        outp = outp.split()
        network[inpu] = outp

    return network

def count_output(network, start='you', end=['out'], ignore=[]):
    path = [start]
    queue = deque([[start, deepcopy(path)]])

    count = {}
    for e in end:
        count[e] = 0
    
    while queue:
        nxt = queue.popleft()
        #print(nxt)
        node, path = nxt

        connections = network[node]
        for conn in connections:
            if conn in end:
                count[conn] += 1
            elif conn in ignore:
                pass
            elif conn not in path:
                queue.append([conn, path+[conn]])

    return count

def run_part(data, part=1):
    network = build_network(data)

    if part == 1:
        return count_output(network)['out']

    if part == 2:
        print('svr -> dac/fft')
        count_svr_dacfft = count_output(network, start='svr', end=['dac', 'fft'], ignore=['out'])
        print(count_svr_dacfft)
        print('dac -> fft')
        count_dac_fft = count_output(network, start='dac', end=['fft'], ignore=['out', 'svr'])
        print('fft -> dac')
        count_fft_dac = count_output(network, start='fft', end=['dac'], ignore=['out', 'svr'])
        print('dac -> out')
        count_dac_out = count_output(network, start='dac', end=['out'], ignore=['fft', 'svr'])
        print('fft -> out')
        count_fft_out = count_output(network, start='fft', end=['out'], ignore=['dac', 'svr'])

        count = count_svr_dacfft['dac']*count_dac_fft['fft']*count_fft_out['out']
        count += count_svr_dacfft['fft']*count_fft_dac['dac']*count_dac_out['out']
        return count

    return
```

```python
print('Part 2 test:', run_part(test2, 2), run_part(test2, 2)==2)
```

```python
with open('input_day11.txt', 'r') as f:
    data = f.readlines()
    f.close()

print('Part 2 result:', run_part(data, 2))
```

```python
print('Part 1 test:', run_part(test), run_part(test)==5)
```
