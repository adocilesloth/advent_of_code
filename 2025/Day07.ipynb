```python
test = ['.......S.......',
        '...............',
        '.......^.......',
        '...............',
        '......^.^......',
        '...............',
        '.....^.^.^.....',
        '...............',
        '....^.^...^....',
        '...............',
        '...^.^...^.^...',
        '...............',
        '..^...^.....^..',
        '...............',
        '.^.^.^.^.^...^.',
        '...............']
```

```python
def parse_data(data):
    graph = []
    for line in data:
        graph.append(list(line.strip()))
    return graph

def count_splits(graph):
    queue = []
    
    #find start
    for y in range(0, len(graph[0])):
        if graph[0][y] == 'S':
            queue.append((0,y))
            
    splitters = []
    #beams
    while(len(queue)):
        node = queue.pop(0)
        x = node[0]
        y = node[1]
        
        if x == 0:
            dys = [0]
        else:
            dys = [-1,1]
        
        for dy in dys:
            for i in range(x, len(graph)):
                if graph[i][y+dy] == '^':
                    if (i,y+dy) not in splitters:
                        splitters.append((i,y+dy))
                        queue.append((i,y+dy))
                    break
                elif graph[i][y+dy] != 'S':
                    graph[i][y+dy] = '|'
                    
    return len(splitters)

def next_split(graph, start, count=0, memory={}):
    split = False
    for x in range(start[0], len(graph)):
        if graph[x][start[1]] == '^':
            split = True
            if (x,start[1]) in memory.keys():
                return count+memory[(x, start[1])]
            break
            
    if split == False:
        return count+1
    
    left_count = next_split(graph, (x, start[1]-1), count, memory)
    right_count = next_split(graph, (x, start[1]+1), left_count, memory)
    
    memory[(x, start[1])] = right_count - count
    
    return right_count
    

def count_paths(graph): #this is DFS with memorisation
    #find start
    for y in range(0, len(graph[0])):
        if graph[0][y] == 'S':
            start = (0,y)
            
    count = next_split(graph, start)
    return count
    
    
def run_day(data, part=1):
    graph = parse_data(data)
    
    if part == 1:
        result = count_splits(graph)
    elif part == 2:
        result = count_paths(graph)
        
    return result
```

```python
print('Part 1 test:', run_day(test), run_day(test)==21)
```
