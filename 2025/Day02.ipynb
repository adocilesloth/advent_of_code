```python
def find_repints(sequence):
    limits = sequence.split(',')

    repints = []

    for limit in limits:
        lower, higher = limit.strip().split('-')
        low = int(lower)
        upp = int(higher)

        max_len = len(higher)//2
        delta_len = len(higher)-len(lower)

        for i in range(1, max_len+1):
            i_low = min([int(lower[:i]), int(higher[:i])])
            i_upp = int(higher[:i+delta_len])

            for stub in range(i_low, i_upp+1):
                new_int = str(stub)
                while True:
                    if int(new_int+str(stub)) > upp:
                        break
                    new_int += str(stub)
                    
                if new_int == str(stub):
                    continue
                
                new_int = int(new_int)
                if new_int >= low and new_int <= upp and new_int not in repints:
                    repints.append(new_int)

    return repints

def find_double_repints(sequence):
    repints = find_repints(sequence)
    double_repints = []

    for repint in repints:
        strint = str(repint)
        if len(strint)%2 == 1:
            continue

        first_stub = strint[:len(strint)//2]
        last_stub = strint[len(strint)//2:]

        if first_stub == last_stub:
            double_repints.append(repint)

    return double_repints
```

```python
double_repints = find_double_repints(test)
print('Part 1 test:', sum(double_repints), sum(double_repints)==1227775554)
```
